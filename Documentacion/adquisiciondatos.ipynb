```python
import pandas as pd
from sodapy import Socrata
import os
from dotenv import load_dotenv

load_dotenv('.env')

TOKEN_APP = os.getenv("TOKEN_SODAPY")
socrata_dataset_identifier = os.getenv("DATASET_ID")

client = Socrata("www.datos.gov.co", TOKEN_APP)
```

```python
client.timeout = 10000
```

```python

consulta = """SELECT * 
WHERE nombre_entidad ='IDRD - ENTIDAD OFICIAL'
LIMIT 30000
"""
```

```python
results = client.get("jbjy-vk9h", query = consulta)

results_df = pd.DataFrame.from_records(results)
```
